### Análisis Univariado

```python
from matplotlib import style
style.use('ggplot')

# Crear histograma de longitud de pétalo
plt.hist(iris["petal_length"], bins=15, color='c', alpha=0.5)

# Añadir título al gráfico y etiquetas de los ejes
plt.title('Histograma de Longitud de Pétalo', fontsize=14)
plt.xlabel('Longitud de Pétalo', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Añadir líneas verticales para mostrar la media y la mediana
plt.axvline(iris["petal_length"].mean(), color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(iris["petal_length"].median(), color='k', linestyle='dashed', linewidth=2, label='Mediana')

# Añadir leyenda al gráfico
plt.legend(fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
from matplotlib import style
style.use('tableau-colorblind10')

plt.hist(iris["petal_length"])
plt.title('Histograma de Longitud de Petalo', alpha=0.5)
plt.axvline(iris["petal_length"].mean(), color='b')
plt.axvline(iris["petal_length"].median(), color='k')
plt.show()
```

#### Histograma de Ancho de Pétalo:
La distribución de los datos no sigue una distribución normal. Aunque la diferencia entre la media (1.199) y la mediana (1.3) no es considerable, se llevará a cabo un análisis de sesgo para evaluar si la distribución está sesgada hacia algún lado.

Calculando la medida de asimetría A = |min – media| = |0.100 – 1.199| = 1.099 y la medida de apuntamiento B = |max – mediana| = |2.500 – 1.3| = 1.2, se encuentra que A es mayor que B.

Como la diferencia entre A y B es significativa y A es mayor que B, podemos concluir que la distribución está sesgada hacia la izquierda. Esto significa que hay más valores en el extremo inferior de la distribución que en el superior. Este análisis es importante para entender mejor la distribución de los datos y las posibles implicaciones para futuros análisis o decisiones basadas en estos datos.

```python
style.use('ggplot')
plt.hist(iris["petal_width"])
plt.title('Histograma de Anchura de Petalo', alpha=0.5)
plt.axvline(iris["petal_width"].mean(), color='b')
plt.axvline(iris["petal_width"].median(), color='k')
plt.show()
```

#### Histograma de Longitud de Sépalo.
Al evaluar la distribución de los datos, se puede observar que la distancia entre la media y la mediana no es muy grande. En otras palabras, la mediana no se aleja significativamente de la media, lo que sugiere que la distribución podría ser simétrica. Además, al graficar los datos en un histograma o un gráfico de densidad, se puede notar que los datos tienen una forma aproximadamente normal.

A pesar de que la apariencia visual de los datos sugiere que la distribución es normal, es importante llevar a cabo un análisis más detallado para evaluar la presencia de sesgo

```python
plt.hist(iris["sepal_length"])
plt.title('Histograma de longitus de setalo', alpha=0.5)
plt.axvline(iris["sepal_length"].mean(), color='b')
plt.axvline(iris["sepal_length"].median(), color='k')
plt.show()
```

#### Histograma de Anchura de Sépalo.

### Diagramas de Caja

```python
# Seleccionar las variables numéricas para el gráfico
numerical_vars = iris.columns[:-1]

# Crear una figura y una cuadrícula de subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4), sharey=True)

# Dibujar un boxplot en cada subplot
for i, var in enumerate(numerical_vars):
    sns.boxplot(y=var, data=iris, ax=axes[i], color='c')
    axes[i].set_xlabel(None)
    axes[i].set_title(var.title())

# Ajustar los espacios entre subplots y agregar etiquetas de los ejes
fig.tight_layout()
fig.text(0.5, -0.02, 'Variable', ha='center', fontsize=12)


# Mostrar el gráfico
plt.show()
```

```python
# Crear una figura y una cuadrícula de subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

# Dibujar un boxplot en cada subplot para cada especie
for i, var in enumerate(iris.columns[:-1]):
    sns.boxplot(x="species", y=var, data=iris, ax=axes[i], palette="pastel")
    axes[i].set_xlabel(None)
    axes[i].set_title(var.title())

# Ajustar los espacios entre subplots y agregar etiquetas de los ejes
fig.tight_layout()
fig.text(0.5, -0.02, 'Especie', ha='center', fontsize=12)
fig.text(-0.02, 0.5, 'Valor', va='center', rotation='vertical', fontsize=12)

# Mostrar el gráfico
plt.show()
```

## Correlación

Esta es una matriz de correlación que muestra la correlación entre las cuatro variables numéricas del dataset iris: longitud y ancho del sépalo y longitud y ancho del pétalo. Cada correlación se mide en una escala de -1 a 1, donde -1 indica una correlación negativa perfecta, 0 indica que no hay correlación y 1 indica una correlación positiva perfecta. Observamos una alta correlación positiva entre la longitud del sépalo y la longitud del pétalo, así como entre la longitud y el ancho del pétalo. La correlación entre el ancho del sépalo y la longitud del pétalo es moderada y negativa, mientras que la correlación entre el ancho del sépalo y el ancho del pétalo es débil y negativa

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
# Crear un heatmap con etiquetas para cada variable
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')

# Ajustar las etiquetas y leyendas del gráfico
plt.title('Matriz de correlación de las variables numéricas en el dataset iris')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Mostrar el gráfico
plt.show()
```

```python
# Calcular la matriz de correlación
corr = iris.corr()

# Crear una matriz de ceros del mismo tamaño que la matriz de correlación
mask = np.zeros_like(corr)

# Crear una máscara triangular inferior para mostrar solo la mitad inferior de la matriz de correlación
mask[np.triu_indices_from(mask)] = True

# Crear un heatmap con etiquetas para cada variable y la máscara triangular
sns.heatmap(corr, annot=True, cmap='coolwarm', mask=mask)

# Ajustar las etiquetas y leyendas del gráfico
plt.title('Matriz de correlación de las variables numéricas en el dataset iris')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Mostrar el gráfico
plt.show()
```

```python
from scipy.stats import kendalltau
# Cargar el dataset tips
tips = sns.load_dataset('tips')

# Calcular la correlación de Kendall entre las variables total_bill y tip
corr, p_value = kendalltau(tips['total_bill'], tips['tip'])

# Mostrar el resultado
print('Correlación de Kendall: {:.2f}'.format(corr))
print('p-value: {:.4f}'.format(p_value))
```

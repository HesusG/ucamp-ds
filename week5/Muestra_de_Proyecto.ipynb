```python
# Imprimir matriz de correlación
corr_matrix = df_iris.corr()
print(corr_matrix)
```

```python
# Calcular los valores atípicos por variable
Q1 = df_iris.quantile(0.25)
Q3 = df_iris.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_iris < (Q1 - 1.5 * IQR)) | (df_iris > (Q3 + 1.5 * IQR))).sum()

# Calcular el porcentaje de valores ausentes por variable
missing_values = df_iris.isna().sum()
percent_missing = (missing_values / len(df_iris)) * 100

# Imprimir los valores atípicos y el porcentaje de valores ausentes por variable
print("Valores atípicos por variable:")
print(outliers)
print("\nPorcentaje de valores ausentes por variable:")
print(percent_missing)
```

## Correlación de las variables <a name="correlacion"></a>

```python
# Calcular la matriz de correlación
corr_matrix = df_iris.corr()

# Crear una máscara para ocultar la mitad inferior de la matriz
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Personalizar la figura
fig, ax = plt.subplots(figsize=(8, 6))
plt.title('Matriz de correlación', fontsize=16)

# Visualizar la matriz de correlación con una paleta de colores apropiada
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f',
            vmin=-1, vmax=1, center=0, square=True, linewidths=.5,
            cbar_kws={"shrink": .5}, mask=mask)

# Ajustar el espacio de la figura
plt.subplots_adjust(top=0.92, bottom=0.15, left=0.15, right=0.95)

plt.show()
```

La correlación de Pearson es un método estadístico utilizado para medir la fuerza y la dirección de la relación lineal entre dos variables continuas. El coeficiente de correlación de Pearson varía de -1 a 1, donde -1 indica una correlación negativa perfecta, 0 indica que no hay correlación y 1 indica una correlación positiva perfecta.

La correlación de Pearson es ampliamente utilizada en la ciencia de datos y el análisis estadístico debido a su simplicidad y eficacia para medir la relación lineal entre variables continuas. Sin embargo, es importante tener en cuenta que la correlación de Pearson solo mide la relación lineal entre variables y no es adecuada para medir relaciones no lineales.

Por otro lado, la correlación de Pearson no es adecuada para medir la relación entre variables categóricas, ya que estas variables no tienen un orden natural y la correlación de Pearson supone una relación lineal entre las variables.

Además, la correlación de Pearson no es adecuada para medir la relación entre variables no continuas, ya que esta medida de correlación se basa en la covarianza de las variables y supone una relación lineal entre ellas. Para variables no continuas o discretas, es más adecuada la correlación de Kendall o la prueba de chi cuadrado.

## Gráfica de Barras e Histograma <a name="graficas"></a>

```python
# Crear gráfico de barras para la longitud promedio de los pétalos por especie
sns.barplot(data=df_iris, x='species', y='petal_length', ci=None)
plt.title('Longitud promedio de los pétalos por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del pétalo (cm)')
plt.show()
```

```python
# Crear gráfico de barras para la anchura promedio de los sépalos por especie
sns.barplot(data=df_iris, x='species', y='sepal_width', ci=None)
plt.title('Anchura promedio de los sépalos por especie')
plt.xlabel('Especie')
plt.ylabel('Anchura del sépalo (cm)')
plt.show()
```

```python
# Obtener la cantidad de cada especie en el conjunto de datos
species_counts = df_iris['species'].value_counts()

# Crear gráfico de barras para la cantidad de cada especie de iris en el conjunto de datos
sns.barplot(x=species_counts.index, y=species_counts.values)
plt.title('Cantidad de cada especie de iris en el conjunto de datos')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.show()
```

```python
sns.histplot(data=df_iris, x='sepal_length')
plt.title('Histograma de la longitud del sépalo')
plt.xlabel('Longitud del sépalo (cm)')
plt.show()
```

```python
# Crear histograma de la longitud del pétalo por especie
g = sns.FacetGrid(df_iris, col='species', height=6) 
g.map(sns.histplot, 'petal_length')
g.set_titles('{col_name}')
plt.suptitle('Histograma de la longitud del pétalo por especie')
plt.xlabel('Longitud del pétalo (cm)')
plt.tight_layout() # Ajustar la posición de los subplots
plt.show()
```
